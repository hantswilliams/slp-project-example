```python
### recoding for questions 7, 9, 10

recode_dict_v1 = {
    1: 3,
    2: 1,
    3: 2
}

recode_dict_v1
```

```python
## use recode_dict_v1 for Q7, Q9, Q10

df_simple_recode = df_simple.copy()

df_simple_recode["Q7"] = df_simple_recode["Q7"].map(recode_dict_v1)
df_simple_recode["Q9"] = df_simple_recode["Q9"].map(recode_dict_v1)
df_simple_recode["Q10"] = df_simple_recode["Q10"].map(recode_dict_v1)

df_simple_recode
```

```python
### recoding for questions 11

recode_dict_v2 = {
    1: 1,
    4: 1,
    5: 1,
    2: 2,
    3: 2
}

recode_dict_v2
```

```python
## use recode_dict_v1 for Q11

df_simple_recode["Q11"] = df_simple_recode["Q11"].map(recode_dict_v2)

df_simple_recode
```

```python
df_simple_recode['total_score'] = df_simple_recode.sum(axis=1)
df_simple_recode
```

```python
# q2, 4, 5,12 13 - summation for environment_satisfaction

df_simple_recode['environment_satisfaction'] = df_simple_recode['Q2'] + df_simple_recode['Q4'] + df_simple_recode['Q5'] + df_simple_recode['Q12'] + df_simple_recode['Q13_1'] + df_simple_recode['Q13_2'] + df_simple_recode['Q13_3'] + df_simple_recode['Q13_4']

df_simple_recode
```

```python
# q8 and q10 for routine_satisfaction

df_simple_recode['routine_satisfaction'] = df_simple_recode['Q8'] + df_simple_recode['Q10']

df_simple_recode
```

```python
# 1, 2, 6, 9, 10 for training_quality_satisfaction

df_simple_recode['training_quality_satisfaction'] = df_simple_recode['Q1'] + df_simple_recode['Q2'] + df_simple_recode['Q6'] + df_simple_recode['Q9'] + df_simple_recode['Q10']

df_simple_recode
```

```python
df_simple_scores = df_simple_recode[['total_score', 'environment_satisfaction', 'routine_satisfaction', 'training_quality_satisfaction']]

df_simple_scores
```
